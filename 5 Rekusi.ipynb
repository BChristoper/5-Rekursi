```python
#1 Fibonacci

def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(10)) 
```

```python
#2 Faktorial
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)


print(factorial(6)) 
```

```python
#3 Bilangan Alami
def jumlah_bilangan_alami(n):
    if n == 1:
        return 1
    else:
        return n + jumlah_bilangan_alami(n - 1)


print(jumlah_bilangan_alami(20)) 
```

```python
#4 Pangkat
def pangkat(a, b):
    if b == 0:
        return 1
    else:
        return a * pangkat(a, b - 1)


print(pangkat(2, 10)) 
```

```python
#5 FPB
def fpb(a, b):
    if b == 0:
        return a
    else:
        return fpb(b, a % b)

print(fpb(48, 18))
```

```python
#6 Palindrom
def palindrom(teks):
    if len(teks) <= 1:
        return True
    elif teks[0] != teks[-1]:
        return False
    else:
        return palindrom(teks[1:-1])


print(palindrom("roro"))  
print(palindrom("roror"))  
```
